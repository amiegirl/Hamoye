## Drop irrelevant columns

```python
#drop date column
df.drop(columns=['date'],inplace=True)
df
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn import linear_model
```

```python
#12
simple_linear_reg_df = df[['Living_Room_Temp_Celsuis','Outside_Building_northside_Temp_Celsius']].sample(40,random_state=2)
#regression plot
sns.regplot(x='Living_Room_Temp_Celsuis',y='Outside_Building_northside_Temp_Celsius',data=simple_linear_reg_df)
```

```python
#normalise our data set to a common scale using min max scaler
scaler=MinMaxScaler()
normalised_df=pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
features_df=np.array(normalised_df.Living_Room_Temp_Celsuis).reshape(-1,1)
target_df=normalised_df['Outside_Building_northside_Temp_Celsius']
#now we split our dataset into the training and testing dataset
x_train,x_test,y_train,y_test=train_test_split(features_df,target_df,test_size=0.3,random_state=1)
linear_model=LinearRegression()
#fit the model to the training dataset
linear_model.fit(x_train,y_train)
#obtain predictions
predicted_values=linear_model.predict(x_test)
r2_score=r2_score(y_test,predicted_values)
round(r2_score,2)
```

```python
#13
#normalise our data set to a common scale using min max scaler
scaler=MinMaxScaler()
normalised_df=pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
features_df=normalised_df.drop(columns=['lights','Appliances'])
target_df=normalised_df['Appliances']
#now we split our dataset into the training and testing dataset
x_train,x_test,y_train,y_test=train_test_split(features_df,target_df,test_size=0.3,random_state=42)
mlr_model=LinearRegression()
#fit the model to the training dataset
mlr_model.fit(x_train,y_train)
#obtain predictions
mlr_predicted_values=mlr_model.predict(x_test)
mae=mean_absolute_error(y_test,mlr_predicted_values)
round(mae,2)
```

```python
#14
rss=np.sum(np.square(y_test - mlr_predicted_values))
round(rss,2)
```

```python
#15
rmse=np.sqrt(mean_squared_error(y_test,mlr_predicted_values))
round(rmse,3)
```

```python
#16
mlr_score=r2_score(y_test,mlr_predicted_values)
round(mlr_score,2)
```

```python
#17
#get weights
def get_weights_df(model,feat,col_name):
  weights=pd.Series(model.coef_,feat.columns).sort_values()
  weights_df=pd.DataFrame(weights).reset_index()
  weights_df.columns=['Features',col_name]
  weights_df[col_name].round(3)
  return weights_df
```

```python
#create weight
linear_model_weights_df=get_weights_df(mlr_model,x_train,'Linear_model_weight')
linear_model_weights_df
```

```python
#18
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=0.4)
ridge_reg.fit(x_train,y_train)
ridge_predictions=ridge_reg.predict(x_test)
rmse1=np.sqrt(mean_squared_error(y_test,ridge_predictions))
round(rmse1,3)
```

```python
#19
from sklearn.linear_model import Lasso
lasso_reg= Lasso(alpha=0.001)
lasso_reg.fit(x_train,y_train)
lasso_predictions=ridge_reg.predict(x_test)
lasso_weights_df=get_weights_df(lasso_reg,x_train,'Lasso_weight')
lasso_weights_df
```

```python
#20
rmse2=np.sqrt(mean_squared_error(y_test,lasso_predictions))
round(rmse2,3)
```
